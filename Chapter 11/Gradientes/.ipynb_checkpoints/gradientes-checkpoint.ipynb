<h2 style='font-size:30px'> 
    <a href='https://www.youtube.com/watch?v=0iNrGpwZwog'>The Vector Dot Product 
    </a>
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O produto escalar (dot product) entre dois vetores consiste na soma do produto dos elementos de mesma posição.
        </li>
        <li> 
            Por exemplo, o produto escalar de $a=\begin{bmatrix}a_1 \\ a_2 \end{bmatrix}$ e $b=\begin{bmatrix} b_1 \\ b_2 \end{bmatrix}$  é:
            <p style='margin-top:10px'>
            $$
               a \cdot b = a_1 b_1 + a_2 b_2 
            $$
            </p>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Propriedades do Dot Product</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            $a \cdot b = b \cdot a$
        </li>
        <li> 
            $a\cdot(b+c)=b\cdot(a+c)=c\cdot(a+b)$           
        </li>
        <li> 
            $k(a\cdot b)=a\cdot(kb)=b\cdot(ka)$ k é uma constante
        </li>
        <li> 
            $a\cdot a=|a|^{2}$
        </li>
        <li> 
            $0 \cdot a = \vec{0}$
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Fórmula Trigonométrica do Produto Escalar</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O dot product também pode ser medido como:
            $$a \cdot b = |a||b|\cos{\theta}$$
        </li>
        <li style='margin-top:10px'> 
            Caso o produto escalar seja zero, dizemos que os vetores são ortogonais.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Exemplo 1</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Meça o ângulo entre os vetores $a=\begin{bmatrix} 2 \\ 2 \\-1 \end{bmatrix}$ e $b=\begin{bmatrix}5 \\ -3 \\2 \end{bmatrix}$
            $$a \cdot b = |a||b|\cos{\theta}$$
        </li>
    </ul>
</div>

```python
import numpy as np
a,b = np.array([2,2,-1]), np.array([5,-3,2])

# Medindo o coseno com a fórmula do dot product. A magnitude de um vetor pode ser medida com 'np.linalg.norm'.
cos = np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Obtendo o ângulo;
np.arccos(cos)
```

<h3 style='font-size:30px;font-style:italic'> Exemplo 2</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Tendo $a=\begin{bmatrix} 4 \\ -2 \\ 5\end{bmatrix}$, $b=\begin{bmatrix} -1 \\ 3 \\ -6\end{bmatrix}$ e $c=\begin{bmatrix} 7 \\-5 \\ 1 \end{bmatrix}$.
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    Meça $a \cdot b$
                </li>
                <li> 
                    Meça $b \cdot c$
                </li>
            </ul>
        </li>
    </ul>
</div>

```python
a = np.array([4, -2, 5])
b = np.array([-1, 3, -6])
c = np.array([7, -5, 1])

# A.
print(f'A: {np.dot(a, b)}')

# B
print(f'B: {np.dot(b,c)}')
```

<h2 style='font-size:30px'> O que é um gradiente?</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O gradiente consiste em um valor que indica a variação de uma função de múltiplas variáveis em um certo ponto no espaço. É representado como um vetor que contém as derivadas parciais da função para cada input.
        </li>
        <li> 
            Tomemos como exemplo a função $f(x,y)=x^{2}\sin{y}$. Seu gradiente seria representado como 
            $$
            \nabla f(x,y)=
            \begin{bmatrix}
                \frac{\delta f}{\delta x} \\
                \frac{\delta f}{\delta y}
            \end{bmatrix}
            =
            \begin{bmatrix}
                2x\sin{y}\\
                x^{2}\cos{y}
            \end{bmatrix}
            $$
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> <a href='https://tutorial.math.lamar.edu/classes/calciii/directionalderiv.aspx'>As derivadas direcionais
    </a>
</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Uma derivada direcional busca computar a taxa de variação instantânea de uma função de múltiplas variáveis. Por possuir mais de um argumento, a função pode variar de diversas maneiras a depender de como cada variável tem o seu valor mudado.
        </li>
        <li> 
            Por exemplo, para $f(x,y)$, é possível computar a variável direcional quando x varia positivamente e y negativamente, ou vice-versa. Cada situação gerará um resultado distinto.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Medindo uma variável direcional</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Para descobrirmos a derivada direcional da função f(x,y) na direção do vetor $\vec{v}=\begin{bmatrix}a\\ b \end{bmatrix}$, devemos nos certificar de que a magnitude desse mesmo seja 1 ($||\vec{v}||=1$). Às vezes, a direção pode ser dada por um ângulo, sendo  $\vec{v}=\begin{bmatrix}\cos{\theta} \\ \sin{\theta} \end{bmatrix}$.
        </li>
        <li> 
                    A magnitude de um vetor é dada por $||\vec{v}||=\sqrt{a^{2}+b^{2}+c^{2}}$. Para vetores bidirecionais, desconsidere o c.
        </li>
        <li> 
            A derivada direcional será o produto escalar (dot product) entre o gradiente da função no ponto escolhido e o vetor de direção:
            $$
                D_{\vec{v}}f(x,y)=\nabla{f(x,y)} \vec{v}
            $$
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Exemplo</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Meça a Derivada Direcional de $f(x,y)=xe^{xy}+y$ no ponto (2,0) em que o vetor $\vec{v}$ tenha a direção de $\frac{2\theta}{3}$.
            <ul style='list-style-type:lower-alpha'> 
            </ul>
        </li>
    </ul>
</div>

```python
import numpy as np
from sympy import cos, sin, diff, pi
from sympy.abc import x,y

# Função do exercício.
fxy = x*np.e ** (x*y) + y

# Gradiente no ponto dado.
nabla = np.array([diff(fxy, x).subs({x:2, y:0}), diff(fxy, y).subs({x:2, y:0})], dtype=np.float64)

# Vetor direcional.
direction = np.array([cos(2*pi/3), sin(2*pi/3)], dtype=np.float64)

# Medindo a derivada direcional. Lembre-se de SOMAR os valores do vetor resultante.
print(np.sum(nabla * direction))

# O Dot Produt pode ser obtido também com 'np.dot'.
np.dot(nabla, direction)
```

<h2 style='font-size:30px'>
    <a href='https://www.youtube.com/watch?v=b-RdQn2gQ70'>
    Steepest Ascent and Steepest Descent
    </a>
</h2>

<div> 
    <ul style='font-size:20px'> 
        <li> 
            Existem situações nas quais queremos certa estratégia para alcançarmos um determinado valor de uma função. No contexto do cálculo mutilvarial, conseguimos usar o gradiente como um indicador da direção por onde a equação terá o seu maior acréscimo/decréscimo. 
        </li>
        <li> 
            Como a Derivada Direcional de uma função consiste em um produto escalar, podemos escrever a sua fórmula também da seguinte maneira:
            $$
                D_\vec{v}f(x,y)=|\nabla f(x,y)| |\vec{v}|  \cos{\theta}
            $$
        </li>

<h3 style='font-size:30px;font-style:italic'> Medindo a Maior/Menor Derivada Direcional</h3>
<div>        
    <ul style='font-size:20px'>
        <li> 
            Como $\nabla f(x,y)$ em um ponto sempre terá um único valor e $|\vec{v}|=1$, o único argumento que pode ser mudado é $\theta$. Se quisermos a maior derivada direcional, teremos que iguala-lo a $0^{\circ}$; no caso do maior decréscimo, theta tem que ser $180 ^{\circ}$. Ou seja, $\vec{v}$ sempre terá o mesmo sentido que seu gradiente nesses cenários, podendo ser considerado como $v=\frac{\nabla f}{|\nabla f|}$.
        </li>
        <li> 
            Com isso, no contexto da subida, a maior variação da função terá o valor de: 
            $$
                D_{\vec{v}}f=\nabla f \cdot \frac{\nabla f}{|\nabla f|}=\frac{|\nabla f|^{2}}{|\nabla f|}=|\nabla f|
            $$
        </li>
        <li style='margin-top:10px'> 
            Na descida de gradiente $D_{\vec{v}}f=-|\nabla f|$
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Exemplo 1</h3>
<div>        
    <ul style='font-size:20px'>
        <li> 
            Sendo $f(x,y)=x^{2}y+e^{xy}\sin{y}$ em $(x_0, y_0)=(1,0)$:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    Em qual direção f(x,y) aumenta mais rápido? Qual a derivada dessa direção?
                </li>
                <li> 
                    Em qual direção f(x,y) diminui mais rápido? 
                </li>
            </ul>
        </li>
    </ul>
</div>

```python
from sympy.abc import x,y
from sympy import sin, diff
fxy = x ** 2 * y + np.e ** (x * y) * sin(y)

# Obtendo as derivadas parciais;
nabla = np.array([diff(fxy, l) for l in [x,y] ])

# Medindo o vetor de gradiente no ponto dado.
nabla = np.array([partial.subs({x:1, y:0}) for partial in nabla], dtype='float')

# A.
print(f'A: Na direção: {nabla}. A derivada será: {np.linalg.norm(nabla)}')

# B.
print(f'Na direção:{-nabla}')
```

<h3 style='font-size:30px;font-style:italic'> Exemplo 2</h3>
<div>        
    <ul style='font-size:20px'>
        <li> 
            Existe uma direção pela qual a taxa de variação da temperatura da função $T(x,y,z)=2xy-yz$ é $-3^{\circ}C$ no ponto P(1,-1,1)?
        </li>
    </ul>
</div>

```python
from sympy.abc import x,y,z, theta
from sympy import diff, Eq, solve, cos
fxyz = 2*x*y - y*z
nabla = np.array([diff(fxyz, l).subs({x:1, y:-1, z:1}) for l in [x,y,z]], dtype='float')

# Essa equação tem solução?
eq = Eq(np.linalg.norm(nabla)*cos(theta), -3)

# Não há solução real!
solve(eq)
```

<h2 style='font-size:30px'>
    Aplicação em Machine Learning
</h2>
<div>        
    <ul style='font-size:20px'>
        <li> 
            Se o MSE de uma Regressão Linear é: $MSE(a,b)=(a+2)^{2}+(b-4)^{2}$ sendo a e b os seus coeficientes, descubra o valor-ótimo para cada um deles.
        </li>
        <li> 
            $\nabla{MSE(a,b)}=\begin{bmatrix} 2a + 4 \\ 2b -8\end{bmatrix}$
        </li>
    </ul>
</div>

```python
from sympy.abc import a,b
from sympy import diff

# The function and its gradient.
cost =  (a+2)**2 + (b-4)**2
nabla = np.array([diff(cost, l) for l in [a,b]])

# Iteration adjustments.
n_iterations, learning_rate, tolerance = 1000, .05, 1e-4

# Initial arguments
arguments = np.random.random(size=2)
print(f'Initial Arguments: {arguments}')

for i in range(n_iterations):
    # Gradient at the point under analysis.
    nabla_point = np.array([partial.subs({a:arguments[0], b:arguments[1]}) for partial in nabla], dtype='float')
    # Updating the arguments vector according to the product between the current point's gradient times the learning rate.
    arguments -= nabla_point*learning_rate
    if np.linalg.norm(nabla_point)<tolerance:
        print(f'Optimal Adjustment: {arguments}')
        break
```

```python
import matplotlib.pyplot as plt

a = np.linspace(-10, 10, 100)
b = np.linspace(-10, 10, 100)

a, b = np.meshgrid(a, b)
eq =  (a+2)**2 + (b-4)**2 
fig = plt.figure()

ax = fig.gca(projection='3d')

ax.plot_surface(a, b, eq)
plt.title('(a+2)² + (b-4)² ')

plt.show()
```
