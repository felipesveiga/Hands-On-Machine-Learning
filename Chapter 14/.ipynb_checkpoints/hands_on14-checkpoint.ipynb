<h2 style='font-size:30px'> The Architecture of the Visual Cortex</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            As CNN's têm a sua estrutura inspirada no córtex visual. Em estudos, foi constatado que cada neurônio visual é acionado quando há estímulo em uma área específica do espaço - os receptive fields.
        </li>
        <li> 
            Além disso, certos neurônios são incitados com estímulos mais complexos provenientes de receptive fields maiores. Isso levou à ideia de que células nervosas de maior nível são abastecidas por reações de outras de mais baixo nível.
        </li>
        <li> 
            É importante ressaltar que as CNN's possuem camadas parcialmente conectadas, pois imagens costumam ter dimensões maiores do que dados planilhados. Uma fully-conected network requeriaria, portanto, uma quantidade imensa de parâmetros, consumindo muita memória.
        </li>
    </ul>
</div>

<h2 style='font-size:30px'> Convolutional Layers</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            As CNN's têm como seu principal componente as Camadas Convolucionais. Os neurônios dessas não se conectam com todos os outputs da camada anterior, sendo uma tentativa de mimetizar um receptive field. O tamanho desse pode ser livremente escolhido.
        </li>
        <li> 
            Cabe destacar que as camadas de CNN's são em duas - ou três - dimensões, diferentemente das redes vistas até então.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Propriedades das Computações</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Para garantir que a próxima camada tenha o mesmo formato de sua precedente, preenchemos as bordas desta com 0's.
         </li>
    </ul>
</div>

<div> 
    <ul style='font-size:20px;margin-bottom:0px'>
        <li> 
            A fim de reduzirmos a dimensão da camada de output, podemos definir um tamanho maior da passada - stride - de um campo receptivo para outro. Conseguimos escolher o stride vertical e horizontal.
         </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Filters</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Os filtros são as matrizes-kernel que farão a convolução da imagem. Os principais filters são matrizes quadradas com linha ou coluna central de 1's, com o restante de sua estrutura com 0's.
         </li>
        <li> 
            O output de uma camada que usa o mesmo filtro nas convoluções é conhecido como feature map.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Stacking Multiple Feature Maps</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A partir do momento que lidamos com imagens de múltiplos canais (RGB, por exemplo), a natureza das convoluções se altera. Quando desejamos fazer essa operação, devemos recorrer a um kernel também 3-D de mesma profundidade.  
        </li>
        <li> 
            Nós, novamente, multiplicaremos os elementos por posição para somar todos os produtos, mas agora considerando a terceira dimensão. 
        </li>
    </ul>
</div>

<center> 
    <h1> Exemplo de Convolução (Múltiplos Canais)</h1>
</center>

<div> 
    <ul style='font-size:20px'>
        <li> 
            Como as profundidades das matrizes são as mesmas, não podemos deslizar o kernel para frente e trás, fazendo com que o feature map computado seja 2-D. Mas ressalto que isso não é necessariamente uma condição obrigatória para a operação ocorrer.
        </li>
        <li> 
            Esse mesmo cálculo será feito quando tivermos vários feature maps - gerados por uma camada convolucional anterior. Nesse caso, nós os empilhamos e fazemos a convolução com um kernel $K_{(height, width, feature maps)}$.
        </li>
    </ul>
</div>

<h1> Equação da Convolução 3-D</h1>

```python
import numpy as np
import tensorflow as tf
from sklearn.datasets import load_sample_image

flower = load_sample_image('flower.jpg')#load_sample_images()['images'][0].shape
vertical = tf.Variable(tf.zeros(shape=(7, 7, 3), dtype=tf.uint8))
vertical[:, 3, :].assign(1)

# Tentativa de programar uma convolução...
def conv(image:tf.Tensor, kernel:tf.Tensor, hstrides:int=1, wstrides:int=1)->tf.Tensor:
    kernel_shape, image_shape = kernel.shape, image.shape # Armazendo os shapes das matrizes.
    assert image_shape[2] == kernel_shape[2], 'O kernel deve ter mesma profundidade que a matriz' # Verificando a validez da convolução.
    
    # Tensor de output.
    output_h = (image_shape[0] - kernel_shape[0])//hstrides + 1
    output_w = (image_shape[1] - kernel_shape[1])//wstrides + 1
    output = np.empty((output_h, output_w))
    for i in range(kernel_shape[0], image_shape[0]+1, hstrides):
        for j in range(kernel_shape[1], image_shape[1]+1, wstrides):
            output[(i-kernel_shape[0])//hstrides, (j-kernel_shape[1])//wstrides] = tf.reduce_sum(kernel * 
                                                                                image[i-kernel_shape[0]:i, j-kernel_shape[1]:j, :])
    return output
# Consertar iteração.
#convoluted = conv(flower, vertical)
```

<h3 style='font-size:30px;font-style:italic'> TensorFlow Implementation</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Faremos aqui uma Convolução com duas imagens amostrais dentro do TensorFlow.
        </li>
        <li> 
            Na framework, uma imagem é tida como uma matriz de formato $[n_{\text{batches}}, \text{height}, \text{width}, n_\text{channels}]$. Já a matriz de weights, $[\text{height}, \text{width},  n_\text{channels}, n_{\text{filters}}]$
        </li>
    </ul>
</div>

```python
# O método `load_sample_image` nos oferece duas fotos. Uma de um templo chinês, outra de uma flor.
from sklearn.datasets import load_sample_images
import numpy as np
import tensorflow as tf

# Carregando as fotografias.
images = tf.Variable(load_sample_images()['images'])
batch_size, height, width, channels = images.shape # Extraindo os valores de cada dimensão.
```

```python
n_filters = 2 # Faremos as convoluções com 2 filtros.

# Matriz de filtros.
filters = tf.Variable(tf.zeros((7, 7, channels, n_filters), dtype=tf.int32))
filters[3, :, :, 0].assign(1) # Filtro horizontal.
filters[:, 3, :, 1].assign(1) # Filtro vertical.

# Fazendo as convoluções.
outputs = tf.nn.conv2d(input=images, filters=filters, strides=1, padding='SAME', )

output_valid = tf.nn.conv2d(input=images, filters=filters, strides=1, padding='VALID', )
# Nota: `conv2d` admiite apenas inputs integers.
```

```python
import matplotlib.pyplot as plt

plt.imshow(outputs[1, :, :, 0], cmap='gray')
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> padding</h4>
<div> 
    <ul style='font-size:20px'>
        <li> 
            O argumento 'padding' sinaliza se usaremos zero-padding em nossa matriz ou não. Caso passemos "SAME", acrescentaremos zeros nas extremidades da dimensão, garantindo que o kernel tenha se sobreposto a todos os pixels possíveis, sem truncamento. Se usarmos "VALID", não ha acréscimo de zeros e alguns pixels podem ser negligenciados devido ao truncamento do kernel.
        </li>
    </ul>
</div>

```python
# A função com a qual trabalhamos exige que criemos nossos próprios filtros.
# Mas, no contexto da modelagem, eles seriam matrizes inicializadas automaticamente.
import tensorflow.keras as keras

# A `Conv2D` é a camada responsável pelas convoluções de uma CNN
conv = keras.layers.Conv2D(filters=10, kernel_size=3, strides=1, padding='same',
                          activation='relu')

# Realizando a convolução.
conv(tf.cast(images, tf.float32))[0][0]
```

<h2 style='font-size:30px'> Pooling Layers</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            As Pooling Layers têm funcionamento bastante similar ao das convolucionais. Elas também enquadram uma porção do input, mas computam uma operação de agregação sobre os pixels - média, ou máximo. Esse número é o resultado da operação para a parcela da imagem em questão.
        </li>
        <li> 
            A meta é diminuir o tamanho da matriz, poupando processamento e memória. Como a quantidade de coeficientes também cai, elas têm uma certa capacidade de regularização.
        </li>
        <li> 
            A operação é feita em cada canal da matriz. Portanto o output manterá a mesma profundidade.
        </li>
    </ul>
</div>

```python
# Veja que as mesmas configurações das Conv estão presentes nas camadas de Pooling.
pool2d = keras.layers.MaxPool2D(pool_size=(2,2), strides=2, padding='valid')(images)
```

<h3 style='font-size:30px;font-style:italic'> Invariância  de Tradução</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Uma outra vantagem das camadas de Pooling é o oferecimento de traduções invariáveis. Isso quer dizer que um determinado padrão presente em diferentes lugares das fotografias pode ser denunciado no resultado do Pooling.
         </li>
        <li> 
            Observe na imagem abaixo que o output de A e B é igual mesmo com o 8 sendo deslocado. 
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> TensorFlow Implementation</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A aplicação é bastante simples, uma vez que cada tipo de Pooling tem a sua classe no módulo `keras.layers`.
        </li>
        <li> 
            O autor adverte que o Max Pooling costuma dar resultados melhores do que o Average Pooling, além de oferecer uma invariância de traduções mais forte.
        </li>
    </ul>
</div>

```python
# Um breve exemplo do uma Max Pooling.
keras.layers.MaxPool2D(pool_size=2)

# `MaxPooling2D` faz tambem a mesma operação; é apenas um alias.
keras.layers.MaxPooling2D()
```

```python
# É apresentada também a possibilidade de realizarmos Poolings ao longo da profundidade da matriz.
# Como as classes prontas não possibilitam isso, temos que usar as camadas `Lambda` juntamente com os métodos do módulo `tf.nn`.
keras.layers.Lambda(lambda X: tf.nn.max_pool2d(X, ksize=(1,1,1,3)))
```

```python
# Por último, somos apresentados aos Pooling Globais, que computam a operação de agregação sobre todo o feature map, ao invés de em partes dele.
keras.layers.GlobalAvgPool2D()(images)
```

<h3 style='font-size:30px;font-style:italic'> AlexNet</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            É a primeira CNN bem sucedida a realizar convoluções em sequência, não tendo um Pooling intermediário. 
        </li>
    </ul>
</div>

<div> 
    <ul style='font-size:20px'>
        <li> 
            Essa rede usa duas técnicas de regularização: Dropout (rate=50%) entre F8 e F9 e Data Augmentation.
        </li>
        <li> 
            A saída das camadas C1 e C3 sofre a Local Response Normalization, que é a modificação do valor de um pixel com base nas intensidades de células de mesma posição em feature maps vizinhos.
        </li>
    </ul>
</div>

<div> 
    <ul style='font-size:20px'>
        <li> 
            $r, \alpha\text{ e} \beta$ são hiperparâmetros escolhidos pelo usuário. Quando $r=2$, por exemplo, consideramos os pixels dos feature maps de posição $i-\frac{r}{2}$ a $i+\frac{r}{2}$ na soma - lembrando que $f_n -1$ e $0$ são usados caso em caso de valores impróprios.
        </li>
        <li style='margin-top:20px'> 
            No caso da AlexNet, $r=2$, $\alpha=2 \times 10^{-5}$, $\beta=0.75$ e $k=1$.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Data Augmentation</h4>
<div> 
    <ul style='font-size:20px'>
        <li>

        </li>
        <li> 
            O principal objetivo disso é tornar o modelo mais tolerante a variações nas rotações e claridade das imagens. Por isso, é comum girar as fotografias de treino e alterar seu o brilho também.

<h3 style='font-size:30px;font-style:italic'> GoogLeNet (Inception V1)</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A GoogLeNet foi capaz de não apenas aprimorar o desempenho das classificações, como também exigir bem menos parâmetros do que a AlexNet - mesmo sendo mais profunda.
        </li>
        <li> 
            Esta última conquista foi possível graças aos chamados "inception modules", espécies de sub redes neurais. 
        </li>
     </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Inception Module</h4>
<div> 
    <ul style='font-size:20px'>
        <li> 
            A arquitetura dos Inception Modules é representada abaixo. Na legenda, é informado o tamanho da janela de kernel, do stride e o tipo de padding ('S'='same').
        </li>
    </ul>
</div>

<div> 
    <ul style='font-size:20px'>
        <li> 
            Os empilhamentos verticais de Convoluções ou Pooling apresentados são chamados de camadas.
        </li>
        <li> 
            As convoluções 1x1 servem para a captura de padrões ao longo do eixo de profundidade da matriz tridimensional.
        </li>
        <li> 
            Todas as convoluções aplicam ReLU como ativação.
        </li>
    </ul>
</div>

```python
# Os Inception Modules podem ser montados como uma Functional API dentro do Keras.
input_layer = keras.layers.Input(shape=[28, 28, 1])

# O número de filtros de uma camada convolucional é um parâmetro que deve ser ajustado.
# Camada 1.
layer_1 = keras.layers.Conv2D(filters=10, kernel_size=1, strides=1, padding='same')(input_layer)

# Camada 2.
layer_2 = keras.layers.Conv2D(filters=10, kernel_size=1, strides=1, padding='same')(input_layer)
layer_2 = keras.layers.Conv2D(filters=10, kernel_size=3, strides=1, padding='same')(layer_2)

# Camada 3.
layer_3 = keras.layers.Conv2D(filters=10, kernel_size=1, strides=1, padding='same')(input_layer)
layer_3 = keras.layers.Conv2D(filters=10, kernel_size=5, strides=1, padding='same')(layer_3)

# Camada 4.
layer_4 = keras.layers.MaxPool2D(pool_size=3, strides=1, padding='same')(input_layer)
layer_4 = keras.layers.Conv2D(filters=10, kernel_size=1, strides=1, padding='same')(layer_4)

# Concatenando os produtos das 4 camadas.
concat = keras.layers.Concatenate(axis=3)([layer_1, layer_2, layer_3, layer_4])
```

            Todas as camadas convolucionais da arquitetura usam ReLU.

<h3 style='font-size:30px;font-style:italic'> ResNet</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Composta pelas residual units, sequência de camadas convolucionais em que o input de uma camada anterior é somado ao output de uma camada posterior.

<div>

        <li> 
            De quando em quando, usaremos camadas convolucionais de stride 2 com o dobro de filtros. Nessa situação, a matriz transportada pela skip connection não poderá ser somada ao output da Residual Unit.
        </li>
        <li> 
            A solução para esse problema é acrescentar uma camada convolucional dentro da skip com essas mesmas propriedades (kernel_size=1) seguida de uma BN.
        </li>
     </ul>
</div>

<div> 
    <ul style='font-size:20px'>
        <li> 
            Finalmente, segue a arquitetura geral da ResNet. Observe que as linhas curvas representam skip connections, a seta vermelha sinaliza a situação em que aplicamos uma Convolução à matriz da skip:
        </li>
     </ul>

<h3 style='font-size:30px;font-style:italic'> XCeption</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Sua arquitetura tem a peculiaridade de utilizar as depthwise separable convolutions. Elas aplicam um filtro 2-D a cada profundidade do empilhamento de feature maps - resultando em um output de mesma profundidade - seguido por uma convolução 1x1 - essa última a fim de se encontrar padrões entre os pixels de mesma posição.
        </li>
        <li> 
            É importante ressaltar que a primeira sequência de convoluções não é seguida de uma função de ativação.
        </li>
     </ul>
</div>

```python
# No Keras, a primeira operação é feita apenas com um único filtro espacial.
# Portanto, o argumento `filters` do objeto `SeparableConv2D` controla a quantidade de filtros 1x1.
model = keras.models.Sequential([
    keras.layers.Input(shape=[28, 28, 3]),
    keras.layers.SeparableConv2D(filters=129, kernel_size=3,)
])

# Array com os coeficientes da convolução 2-D.
tf.shape(model.layers[0].weights[0])
```

        </li>
        <li> 
            Ao final da rede, fica como opcional a inserção de uma camada densa de 4096 neurônios.
        </li>
     </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> SeNet</h3>
<div> 
    <ul style='font-size:20px'>
        <li> 
              A SeNet é, nada mais, que uma extensão da ResNet ou das Inception Networks. Nela, o produto das unidades residuais ou módulos de incepção passam por uma estrutura conhecida como SE block. 
        </li>
        <li> 
            O bloco é responsável por associar uma relevância a cada feature map. Cada um deles é multiplicado por um float entre 0 e 1. Com isso, feature maps considerados menos essenciais têm os seus valores diminuídos.
        </li>
     </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> SE Block</h4>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Primeiramente, aplicamos uma média global de cada feature map, resultando em um vetor com um float por profundidade.
        </li>
        <li>
            Em seguida, passamos o vetor como input de uma camada densa com ReLU - tipicamente, com 16 vezes menos neurônios. Sua intenção é aprender de maneira generalizada a relação entre os feature maps.
        </li>
        <li> 
            Por fim, uma nova camada densa é usada. Sua quantidade de neurônios é o número de feature maps original; também usará uma sigmóide como ativação. É ela quem computará o grau de relevância de cada feature map.
        </li>
     </ul>
</div>

<h2 style='font-size:30px'> Implementing a ResNet-34 CNN Using Keras</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            Construindo aqui uma ResNet-34 com o auxílio da POO.

    </ul>
</div>

```python
# O Keras não nos providencia uma Residual Unit. Por isso, teremos que programá-la, assim como sugere os diagramas apresentados.
import tensorflow as tf
from tensorflow import keras
from typing import List, Tuple
class ResidualUnit(keras.layers.Layer):
    '''
        Uma ResidualUnit que pode ser implantada dentro de qualquer Rede Convolucional.
        
        Parâmetros
        ----------
        `filters`: int
            Quantidade de filtros a serem aplicados tanto no fluxo padrão, quanto na skip connection.
            
        `kernel_size`: Tuple[int]
            Tamanho das janelas dos filtros.
            
        `strides`: int
            Tamanho da passada dos filtros da primeira camada principal.
            
        `activation`: str
            Função de ativação da primeira camada principal.
    '''
    @staticmethod
    def __main_layers(filters:int, kernel_size:Tuple[int], strides:int, activation:str)->List[keras.layers.Layer]:
        '''
            Construção das camadas principais da ResidualUnit.

            Parâmetros
            ----------
            `filters`: int
                Quantidade de filtros a serem aplicados tanto no fluxo padrão, quanto na skip connection.

            `kernel_size`: Tuple[int]
                Tamanho das janelas dos filtros.

            `strides`: int
                Tamanho da passada dos filtros da primeira camada principal..

            `activation`: str
                Função de ativação da primeira camada principal.
                
            Retorna
            -------
            Uma lista com as camadas principais da arquitetura.
        '''
        return [
                 keras.layers.Conv2D(filters=filters, kernel_size=kernel_size, strides=strides, 
                                      activation=activation, padding='same', use_bias=False),
                 keras.layers.BatchNormalization(),
                 keras.layers.Conv2D(filters=filters, kernel_size=kernel_size, strides=1, padding='same', use_bias=False),
                 keras.layers.BatchNormalization()
               ]
    
    @staticmethod
    def __skip_connection(filters:int, strides:int)->List[keras.layers.Layer]:
        '''
            Construção da Skip Connection da Residual Unit
            
            Parâmetros
            ----------
            `filters`: int
                Quantidade de filtros a serem usados nas convoluções.
                
            `strides`: int
                Tamanho da passada dos filtros na transição de cada convolução.
            
            Retorna
            -------
            Se `strides`==1, a função retorna `None`. Caso o contrário, uma lista contendo uma `Conv2D(filters=filters, strides=strides)` 
            e `BatchNormalization` é dada.
        '''
        if strides == 1:
            return None
        return [
            keras.layers.Conv2D(filters=filters, kernel_size=1, strides=strides, padding='same', use_bias=False),
            keras.layers.BatchNormalization()
        ]
    
    def __init__(self, filters:int, kernel_size:Tuple[int], strides:int, activation:str='relu'):
        self.filters = filters
        self.kernel_size = kernel_size
        self.strides = strides
        self.activation = keras.activations.get(activation)
        self.main_layers = self.__main_layers(filters, kernel_size, strides, activation)
        self.skip_connection = self.__skip_connection(filters, strides)
        
    def main(self, inputs:tf.Tensor)->tf.Tensor:
        '''
            Execução das operações da Unidade.
            
            Parâmetros
            ----------
            `inputs`: `tf.Tensor`
                Tensor com os dados que abastecerão a Unidade Residual.
                
            Retorna
            -------
            Um `tf.Tensor` devidamente transformado.
        '''
        Z = inputs
        # Convoluções na estrutura principal.
        for layer in self.main_layers:
            Z = layer(Z)
        # Convoluções da Skip Connection apenas são acionadas na situação em que `self.stride`>1.
        if self.skip_connection:
            Z_skip = inputs
            for layer in self.skip_connection:
                Z_skip = layer(Z_skip)
            return self.activation(Z+Z_skip)
        return self.activation(Z+inputs)
```

<h2 style='font-size:30px'> Pretrained Models for Transfer Learning</h2>
<div> 
    <ul style='font-size:20px'>
        <li> 
            O Aprendizado por Transferência é adequado quando temos um set de treino enxuto.
        </li>
        <li> 
            Vamos aqui simular um Tranfer Learning com a XCeption.
        </li>
    </ul>
</div>

```python
# Usaremos o modelo na tarefa de distinção de espécies de flores.
# Observe que o dataset contém apenas a porção de treino. Portanto, vamos ter que criar os sets de validação e teste artificialmente.
import tensorflow_datasets as tfds

(train_split, valid_split, test_split), info = tfds.load('tf_flowers', as_supervised=True, split=['train[:75%]', 'train[75%:85%]', 'train[85%:]'], 
                                                     with_info=True)
```

```python
# Ao definir `with_info=True`, ganhamos acesso a algumas informações do dataset.
n_classes = info.features['label'].num_classes
info.features['label'].names
```

```python
from functools import partial
def random_crop(image):
    '''
        Produz recortes aleatórios em nossa imagem.
        
        Parâmetro
        ----------
        `image`: Tensor com os pixels da fotografia.
        
        Retorna
        -------
        A imagem recortada aleatoriamente.
    '''
    shape = tf.shape(image)
    min_dim = tf.reduce_min([shape[0], shape[1]]) * 9 // 10
    return tf.image.random_crop(image, [min_dim, min_dim, 3], seed=42)
    
def preprocess(image, label, randomize=False):
    '''
        Realiza o preprocessamento das imagens para o abastecimento da XCeption.
        
        Parâmetros
        ----------
        `image`: O tensor de pixels da imagem.
        `label`: A categoria da fotografia.
        `randomize`: Booleano indicando se desejamos realizar transformações aleatorizadas.
        
        Retorna
        -------
        Uma tupla com a imagem tratada e a classe.
    '''
    if randomize:
        image = random_crop(image) # Recortando a imagem aleatoriamente.
        image = tf.image.random_flip_left_right(image) # Girando a fotografia horizontalmente.
    image = tf.image.resize(image, [229, 229])
    image = keras.applications.xception.preprocess_input(image)
    return image, label

# Tratando as imagens.
batch_size = 32
train_set = train_split.map(partial(preprocess, randomize=True)).shuffle(500).batch(batch_size).prefetch(1)
valid_set = valid_split.map(preprocess).batch(batch_size).prefetch(1)
test_set = test_split.map(preprocess).batch(batch_size).prefetch(1)
```

```python
# Finalmente, vamos criar o nosso Aprendizado por Transferência caseiro.
xception = keras.applications.xception.Xception(include_top=False)

model_config = xception.get_config()
model_config['layers'] = model_config['layers'][:20] # Preservando apenas as 20 primeiras camadas da XCeption.
model_config['output_layers'] = [['block3_sepconv2_act', 0, 0]] # Definindo a camada de output do modelo-base.

# Criando o modelo a partir das configurações informadas.
base_model = keras.models.Model().from_config(model_config)
base_model.trainable = False # Não permitiremos atualização dos coeficientes do modelo-base.

# Acrescentando algumas camadas novas.
conv2d_0 = keras.layers.Conv2D(50, activation='relu', kernel_size=(3, 3), padding='SAME', strides=1)(base_model.output)
conv2d_1 = keras.layers.Conv2D(50, activation='relu', kernel_size=(3, 3), padding='SAME', strides=1)(conv2d_0)
avg = keras.layers.GlobalAveragePooling2D()(conv2d_1)
output = keras.layers.Dense(n_classes, activation='softmax', use_bias=False, )(avg)

# Finalmente, montando o modelo.
model = keras.models.Model(inputs=base_model.input, outputs=output)
```

```python
# Treinando o modelo.
model.compile(optimizer=keras.optimizers.Adam(learning_rate=.002), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_set, epochs=5, validation_data=valid_set)
```
