<h1 style='font-size:40px'> Outlier detection and removal using IQR</h1>

<div> 
    <ul style='font-size:20px'> 
        <li> 
            Podemos utilizar intervalos interquartis a fim de removermos outliers, sendo IQR = Q3-Q1.
        </li>
        <li> 
        Tipicamente, usa-se o Q1 - 1.5*IQR para eliminação de valores muito baixos e Q3 + 1.5*IQR para se retirar números muito altos.
        </li>
    </ul>
</div>

```python
# Mexeremos aqui com outro dataset sobre alturas.
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/3_outlier_IQR/heights.csv')
df.head()
```

```python
# Veja: nosso dataset possui outliers tanto de pequena, quanto de grande proporção.
df.describe()
```

```python
# Medindo o IQR e as alturas mínima e máxima que iremos admitir.
iqr = df.height.quantile(.75) - df.height.quantile(.25)
min_height = df.height.quantile(.25) - 1.5* iqr
max_height = df.height.quantile(.75) + 1.5 * iqr

# Dataset sem os outliers.
df[(df.height>min_height) & (df.height<max_height)]
```
